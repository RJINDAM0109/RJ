**a)  set a seed (to ensure work reproducibility) and take a random sample of â€¯25 observations and find the mean Glucose and highest Glucose values of this sample and compare these statistics with the population statistics of the same variable. You should use charts for this comparison.**

```
#importing the required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import random
diabetes = pd.read_csv('diabetes.csv')
diabetes.head()
```

```
np.random.seed(1)

# making a sample dataset from the original and calculating mean, max values of glucose
dfn= diabetes.sample(n = 25)
mean_dfn = dfn['Glucose'].mean()
print("The sample dataset's mean glucose value is :",mean_dfn)
highest_dfn = dfn['Glucose'].max()
print("The sample data set's highest glucose value is :",highest_dfn)
```

```
#computing the mean and maximum values of the glucose feature from the entire dataset
ttl_mean = diabetes['Glucose'].mean()
print("The dataset's overall mean value for the glucose attribute is :",ttl_mean)
highest = diabetes['Glucose'].max()
print("The maximum glucose value from the entire dataset is :", highest)
```

```
# plotting the bar graph for mean glucose values calculated above
mean_data = [ttl_mean, mean_dfn]
mean_labels = ['Population', 'Sample']
plt.bar(mean_labels, mean_data)
plt.title('Comparison of Mean Glucose Values')
plt.xlabel('Data Source')
plt.ylabel('Glucose Value')
plt.show()
```

The above bar plot clearly shows that the mean of the sample and population glucose features is almost identical.

```
# plotting the bar graph for max glucose values calculated above
max_data = [highest, highest_dfn]
max_labels = ['Population', 'Sample']
plt.bar(max_labels, max_data)
plt.title('Comparision of Highest Glucose Values')
plt.xlabel('Data Source')
plt.ylabel('Glucose Value')
plt.show()
```

According to the plot above, the greatest value for the sample is about 175 - 200, while the population value is nearly 200. This sample may be close to matching the population findings, but it cannot be accurate for all samples since when the samples change, the values may vary as well.
